Exploring Toronto Neighborhoods – to open a Restaurant

Introduction/Business Problem
Toronto, the capital of the province of Ontario, is the most populous Canadian city. Its diversity is reflected in Toronto’s ethnic neighborhoods such as Chinatown, Corso Italia, Greektown, Kensington Market, Koreatown, Little India, Little Italy, Little Jamaica, Little Portugal & Roncesvalles.
In this project we will go through step by step process to make a decision on a good place to open a restaurant. We analyze the neighborhoods in Toronto to identify the most profitable area by using the income and population of each neighborhoods since the success of the restaurant depends on the people and ambience.
